````python
def absolute_value(x):
  if x < 0:
    return -x
  else:
    return x

absolute_value(-5)
````

````python
def calculate(operation, x, y):
  if operation == 'add':
    return x+y
  elif operation == 'subtract':
    return x-y
  elif operation =='multiply':
    return x*y
  elif operation == 'divide':
    return x/y
  else:
    print(f"Operation {operation} is not defined.")
````

````python
calculate('add',1,2)
````

````python
calculate('addition',1,2)
````
